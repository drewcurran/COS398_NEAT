```python
best_agent = population.species[0].players[0]
game_agents = [NEATPlayer(Color.ORANGE), RandomPlayer(Color.RED), RandomPlayer(Color.BLUE), RandomPlayer(Color.WHITE)]
agent = game_agents[0]
agent.agent = best_agent

play_batch(1000, game_agents)
```

```python
players = []
for species in population.species:
    for player in species.players:
        players.append(player)
players.sort(key=lambda k: k.fitness, reverse=True)

top_players = players[0:100]

play_batch(10, game_agents, population_players=top_players)
```
