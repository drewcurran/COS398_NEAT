```python
with open('../population.pickle', 'rb') as handle:
    population = pickle.load(handle)
best_agent = population.species[0].players[0]
best_agent.nn.print_state()
best_agent.nn.draw_state(numbers=False)
```

```python
with open('../agent_wins.pickle', 'rb') as handle:
    agent_wins = pickle.load(handle)
plt.plot(range(1, len(agent_wins) + 1), agent_wins, color='Orange')
plt.show()

for i in range(len(agent_wins)):
    if agent_wins[i] > 400:
        agent_wins[i] /= 3
plt.plot(range(1, len(agent_wins) + 1), agent_wins, color='Orange')
plt.show()

with open('../agent_wins.pickle', 'wb') as handle:
    pickle.dump(agent_wins, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
